```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4],[10,20,30,40])
plt.title("Sales Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot([1,2,3,4],[10,20,30,40])
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,30,40],'go')
plt.title("Sales Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
x = np.arange(1,5)
y = x**5
plt.plot([1,2,3,4],[10,20,30,40],'go',x,y,'r^')
plt.title('Sales Plot')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
x=np.arange(0,6,0.1)
y = np.sin(x)
plt.plot(x,y,'r')
plt.title('Sales Plot')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)
plt.figure(figsize=(10,5))
plt.plot([1,2,3,4],[10,20,30,40],'go',x,y)
plt.title('Sales Plot')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales = [50,60,55,75,65,80,95,90,85,70,65,55]
plt.bar(month,sales,color='green')
plt.title('bar graph')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
month = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales = [50,60,55,75,65,80,95,90,85,70,65,55]
plt.barh(month,sales,color='green')
plt.title('bar graph')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
