```python
# ========================================================
# PROYECTO: Eficiencia/Automatización de archivos Excel por vendedor
# ========================================================
# Solo debemos cargar el archivo Excel al entorno de Colab
# Cuando abras y utilices este archivo Colab te pedirá que crees una copia en tu entorno (debes aceptar)

import pandas as pd
import os

ruta_base = "/content"
archivo_ventas = os.path.join(ruta_base, "4_Datos.xlsx")
ventas = pd.read_excel(archivo_ventas, sheet_name="datos_resumen")
vendedores = ventas['Vendedor'].unique()

for v in vendedores:
    ventas_filtradas = ventas[ventas['Vendedor'] == v]
    nombre_archivo = v.replace(" ", "_") + ".xlsx"
    ruta_salida = os.path.join(ruta_base, nombre_archivo)
    ventas_filtradas.to_excel(ruta_salida, index=False)
```
